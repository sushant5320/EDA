```python
df_copy['Size'] = df_copy['Size'].replace('Varies with device', np.nan)
```

```python
condlist = [
    df_copy['Size'].str.contains('M'),
    df_copy['Size'].str.contains('k')]

choicelist = [
    df_copy['Size'].str.split('M').str[0],
    df_copy['Size'].str.split('k').str[0]]

df_copy['Size'] = np.select(condlist=condlist, choicelist=choicelist, default=np.nan)
```

```python
size_num = df_copy['Size'].str.extract(r'([\d.]+)').astype(float)
size_unit = df_copy['Size'].str.extract(r'([Mk])')
```

```python
df_copy['Size'] = np.where(
    size_unit[0] == 'M',
    size_num[0] * 1024,
    size_num[0]
).astype('float32')
```

```python
df_copy['Size'] = df_copy['Size'].astype(float).round(2)
```
