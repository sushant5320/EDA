```python
df_copy.drop(df_copy.index[10472], inplace = True)
```

```python
df_copy[~df_copy['Reviews'].str.isnumeric()]
```

```python
df_copy['Size'] = df_copy['Size'].replace('Varies with device', np.nan)
```

```python
size_num = df_copy['Size'].str.extract(r'([\d.]+)').astype(float)
size_unit = df_copy['Size'].str.extract(r'([Mk])')
```

```python
df_copy['Size'] = np.where(
    size_unit[0] == 'k',
    size_num[0] / 1024,
    size_num[0]
).astype('float32')
```

```python
df_copy['Size'] = df_copy['Size'].astype(float).round(3)
```

```python
df_copy.head()
```

```python
df_copy.rename(columns={'Size' : 'Size_in_MB'}, inplace=True)
df_copy.head(2)
```

```python
df_copy[df_copy["Size_in_MB"].isna()]
```

```python
df_copy['Installs'].unique()
```

```python
cond = [
        df_copy['Installs'].str.contains(r'\,'),
        df_copy['Installs'].str.contains(r'\+'),
        df_copy['Price'].str.contains(r'\$')
        ]

choice = [
        df_copy["Installs"].str.replace(',', ''),
        df_copy["Installs"].str.replace('+', ''),
        df_copy["Price"].str.replace('$', '')]

df_copy[['Installs', 'Price']] = np.select(cond, choice, default=df_copy[['Installs', 'Price']])
```

```python
symbol_cols = ['Installs', 'Price']
df_copy[symbol_cols] = df_copy[symbol_cols].replace(r"[+$]", "", regex=True).replace(r"[,]", "", regex=True)
```

```python
df_copy['Installs'] = df_copy['Installs'].astype('int32')
df_copy['Price'] = df_copy['Price'].astype('float32')
```

```python
df_copy.info()
```

```python
df_copy['Last Updated'] = pd.to_datetime(df_copy['Last Updated'])
df_copy["Day"] = df_copy['Last Updated'].dt.day.astype('int16')
df_copy["Month"] = df_copy['Last Updated'].dt.month.astype('int16')
df_copy["Year"] = df_copy['Last Updated'].dt.year.astype('int16')
```
